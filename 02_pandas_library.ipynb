##Pandas Series

```python
#Creating series

import pandas as pd

s1 = pd.Series([10,20,30,40]) #create series from list
s1
```

```python
s2 = pd.Series([10,20,30,40], index = ["a","b","c","d"])
s2
```

```python
#create seried from dictionary

data = {"Ali":55, "Ayse":45,"Mehmet":25}
s3 = pd.Series(data)
s3
```

```python
# Basic Features

s = pd.Series([10,20,30,40],index=["a","b","c","d"])
s
```

```python
print(f"Values: {s.values}")
print(f"Indexes: {s.index}")
print(f"Data type: {s.dtype}")
print(f"Shape: {s.shape}")
```

```python
# Indexing and Slicing

s = pd.Series([10,20,30,40],index=["a","b","c","d"])
s
```

```python
s["b"] #single element access
```

```python
s["a":"c"] #slicing
```

```python
s[0:2] #access with numerical index
```

```python
#functions on series

print(f"Sum: {s.sum()}")
print(f"Mean: {s.mean()}")
print(f"Std: {s.std()}")
print(f"Max: {s.max()}")
print(f"Min: {s.min()}")
```

```python
# unique values

s = pd.Series([10,20,30,30,30,50,50])
s.unique()
```

```python
#the number of each unique value
s.value_counts()
```

## Pandas DataFrames


```python
#creating with dictionaries

import pandas as pd

data = {
    "name":["Ali","Ayse","Mehmet"],
    "age":[25,30,22],
    "city":["Ankara","İstanbul","İzmir"]
}

data
```

```python
df = pd.DataFrame(data)
df
```

```python
#creat with numpy array

import numpy as np

arr = np.array([[1, "Ali", 25],
                [2, "Ayşe", 30],
                [3,"Mehmet", 22]])
```

```python
df2 = pd.DataFrame(arr, columns=["id","name","age"])
df2
```

```python
print(f"Shape (row, column): {df2.shape}")
print(f"Column names: {df2.columns}")
print(f"Row indexes: {df2.index}")
print(f"Data type: \n {df2.dtypes}")
```

```python
# row and column selections

#iloc, loc, indexing

print(df["age"]) #select one column
print(df.loc[:,"age"])
```

```python
print(df[["name","age"]]) #select multiple columns
print(df.loc[:, ["name","age"]])
```

```python
# add and delete column

df["mark"] = [80,90,85]
df
```

```python
df.drop("city", axis=1, inplace = True) # axis = 1 -> delete column
df
```
