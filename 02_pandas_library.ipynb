## Grouping and Summarizing


```python
#Basic grouping

import pandas as pd

data = {
    "department": ["IT","IT","HR","HR","Accounting","IT"],
    "employee": ["ali","ayşe","mehmet","zeynep","burak","can"],
    "salary":[5000,6000,4000,4500,7000,5500]
}

df = pd.DataFrame(data)
df
```

```python
#Grouping by department and average salary

df.groupby("department")["salary"].mean()
```

```python
# total , averaging, and counting

df.groupby("department")["salary"].sum() #total
```

```python
df.groupby("department")["salary"].count() #employee count
```

```python
#min-max
print(df.groupby("department")["salary"].max()) #max salary
print(df.groupby("department")["salary"].min()) #min salary
```

```python
#grouping by multiple columns

data = {
    "department": ["IT","IT","HR","HR","Accounting","IT","HR"],
    "city": ["Ankara","İstanbul","İzmir","Ankara","İstanbul","Ankara","İzmir"],
    "salary":[5000,6000,4000,4500,7000,5500,4700]
}

df = pd.DataFrame(data)
df
```

```python
df.groupby(["department","city"])["salary"].mean()
```

```python
# agg()

df.groupby("department")["salary"].agg(["mean","max","min","count"])
```

## combining and bringing together

```python
#concat

import pandas as pd

df1 = pd.DataFrame({"id":[1,2,3], "name":["Ali","Ayşe","Mehmet"]})
df2 = pd.DataFrame({"id":[4,5,6], "name":["Zeynep","Burak","Elif"]})
```

```python
#add on top of each other

df_concat = pd.concat([df1,df2])
df_concat
```

```python
#add side by side

df_concat = pd.concat([df1,df2], axis = 1)
df_concat
```

```python
# merge
df1 = pd.DataFrame({"id":[1,2,3], "name":["Ali","Ayşe","Mehmet"]})
df2 = pd.DataFrame({"id":[1,2,4], "city":["Ankara","İstanbul","İzmir"]})

print("df1: \n", df1)
print("\n df2:\n",df2)
```

```python
df_merge = pd.merge(df1, df2, on = "id", how="inner")
df_merge
```

```python
df_merge = pd.merge(df1, df2, on = "id", how="left")
df_merge
```

```python
df_merge = pd.merge(df1, df2, on = "id", how="right")
df_merge
```

```python
# index based join

df1 = pd.DataFrame({"name":["Ali","Ayşe","Mehmet"]}, index=["a","b","c"])
df2 = pd.DataFrame({"mark":[85,90,75]}, index=["a","b","d"])


print("df1: \n", df1)
print("\n df2:\n",df2)
```

```python
pd.merge(df1,df2, left_index = True, right_index=True, how="outer") #all values
```
