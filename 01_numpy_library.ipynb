##Statistic and Data Analysis

### mean, median, and mode

```python
print(f"mean: {np.mean(data)}")
```

```python
print(f"median: {np.median(data)}")
```

### variance and standard deviation

```python
print(f"varience : {np.var(data)}")
print(f"std: {np.std(data)}")
```

### Percentile and Quantile

```python
#percentile - Percentile between 0-100

print(f"25. percentile: {np.percentile(data,25)}")
print(f"50. percentile (median): {np.percentile(data,50)}")
print(f"75. percentile: {np.percentile(data,75)}")
```

```python
#quantile - Ratio between 0 and 1

print(f"0.25 quantile: {np.quantile(data,0.25)}")
print(f"0.5 quantile (median): {np.quantile(data,0.5)}")
print(f"0.75 quantile: {np.quantile(data,0.75)}")
```

### correlation and covariance


```python
x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])
```

```python
print(f"correlation: \n {np.corrcoef(x,y)}") # covarience / std
# 1 -> positive relation
#-1 -> negative relation
# 0 -> no relation
```

```python
print(f"covariance: \n {np.cov(x,y)}")
#positive value - increase together
#negative value- decrease together
#0 - no relation
```

### Linear Algebra

```python
#Scalar product

a = np.array([1,2,3])
b = np.array([4,5,6])

print(f"dot: {np.dot(a,b)}")# (1*4)+(2*5)+(3*6)
print(f"vdot: {np.vdot(a,b)}")
```

```python
# matrix product

a = np.array([[1,2],[3,4]])
b = np.array([[5,6],[7,8]])

print(f"Matrix Multiplication: \n {np.matmul(a,b)}")
```

```python
#Determinant

a = np.array([[1,2],[3,4]])
det = np.linalg.det(a)
det
```

```python
# Inverse matrix

a = np.array([[1,2],[3,4]])
inv_a = np.linalg.inv(a)
inv_a
```

```python
#rank

a = np.array([[1,2],[3,4]])
print(f"rank: {np.linalg.matrix_rank(a)}")
```

```python
# Eigen values and vectors

a = np.array([[1,2],[3,4]])
eigenvalues, eigenvectors = np.linalg.eig(a)
print(f"eigenvalues: \n {eigenvalues}")
print(f"eigenvectors: \n {eigenvectors}")
```

```python
#linear equation solution
"""
2x + y = 5
x - y = 1
"""
a = np.array([[2,1],[1,-1]])#coefficients
b = np.array([5,1])

solution = np.linalg.solve(a,b)
solution
```

## masks and conditiaonal operations

```python
#boolean arrays

import numpy as np

arr = np.array([10,20,30,40,50])

mask = arr > 25
print(mask)
print(arr[mask])
```

```python
# Selection based on the "where" command.

arr = np.array([10,20,30,40,50])

result = np.where(arr <25, 0 , arr)
result
```

```python
#select

arr = np.array([10,20,30,40,50])

conditions = [arr <20, arr <40, arr >= 40]
choices = ["small","medium","large"]

labels = np.select(conditions,choices, default = "unknown")
labels
```

```python
# non zero

arr = np.array([10,20,30,40,50])

print(np.nonzero(arr))
```

```python
#any and all

arr = np.array([10,20,30,40,50])

print(f"Is anyone over 30 {np.any(arr > 30)}")
print(f"Are they all greater than 5? {np.all(arr > 5)}")
print(f"Are they all less than 30? {np.all(arr < 30)}")
```
